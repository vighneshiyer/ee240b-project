```python
import SchemDraw as schem
import SchemDraw.elements as e
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
d = schem.Drawing()
vi = d.add(e.DOT_OPEN, label='$V_{in}$')
d.add(e.LINE, l=1, d='right')
ota = d.add(e.OPAMP, anchor='in2', label='OTA', flip=True, d='right')
ota_out = d.add(e.LINE, xy=ota.out, l=0.5)
ota_out_2 = d.add(e.LINE, xy=ota_out.end, d='down', l=2.5)
c3 = d.add(e.CAP, xy=ota_out_2.end, d='down', l=2, label='$C_{2}$')
d.add(e.GND)
d.add(e.LINE, xy=ota_out_2.end, d='left', l=1)
d.add(e.RES, d='left', l=1.1, label='$g_{1}$')
d.add(e.LINE, d='left', l=1.06)
c1 = d.add(e.CAP, d='down', l=2, label='$C_{1}$')
d.add(e.GND)
d.add(e.LINE, xy=ota.in1, d='left', l=0.5)
d.add(e.LINE, d='down', to=c1.start)
d.draw()
d.save('figs/ota_lp_circuit.pdf')
```
